## 1) Le nom d'une antenne a un format bien précis : il est toujours composé de 4 chiffres et de deux lettres. On te demande de retirer tout autre caractère de la colonne "SITE_NAME".

## Surement methode plus pratique mais on va faire en trois temps :

### 1- Garder ce que l'on a apres la premiere parenthese

## Garder ce que l'on a avant la seconde parenthese 

```python
df[['SITE_NAME','SITE2']] = df['SITE_NAME'].str.split(')',expand=True)
```

```python
df
```

## Supprimer les colonnes nouvellement crees 

```python
df = df.drop(columns=['SITE', 'SITE2'])
```

## 2) On te demande de supprimer la colonne "INVOICE_NAME" qui n'apporte rien aux équipes.

## 3) Il y a des antennes pour lesquelles on a des factures mensuelles tandis que pour d'autres, on les reçoit moins régulièrement (trimestre, semestre, année). Ajoute une colonne à ton tableau qui calcule le nombre de jours entre le début de la période de facturation et la fin de la période de facturation.

```python
df
```

## Insertion de la colonne des jours

```python
df.insert(3, "daysPeriod", " ")
```

```python
df
```

### Conversion des dates

```python
df.dtypes
```

```python
df['BEGIN'] = pd.to_datetime(df['BEGIN'], dayfirst=True)
df['END'] = pd.to_datetime(df['END'], dayfirst=True)
```

## Il faut bien ajouter dayfirst = True pour que le format de la date soit respecte

## Soustraction et ajout dans la colonne

```python
df['daysPeriod'] = df['END'] - df['BEGIN']
```

```python
df
```

## 4) A partir de cette nouvelle colonne, crée une autre colonne qui nous donne la consommation journalière moyenne de la période.


### Insertion de la nouvelle colonne

```python
df.insert(5, "kwhPerDay", " ")
```

```python
df
```

### Conversion des jours en integer

```python
df['daysPeriod']
```

```python
# SOURCE : https://stackoverflow.com/questions/25646200/python-convert-timedelta-to-int-in-a-dataframe
```

```python
df['daysPeriod'] = pd.to_numeric(df['daysPeriod'].dt.days, downcast='integer')
```

```python
df['daysPeriod']
```

```python
df.dtypes
```

```python
df
```

### Conversion du KWH_PERIOD en integer

#### Supprimer l'espace dans le nombre

```python
# source : https://stackoverflow.com/questions/43332057/pandas-strip-white-space
```

```python
df['KWH_PERIOD'] = df['KWH_PERIOD'].str.replace(" ","")
```

```python
df
```

```python
# source : https://sparkbyexamples.com/pandas/pandas-convert-column-to-int/
```

```python
df['KWH_PERIOD'] = df['KWH_PERIOD'].astype('int')
```

```python
df.dtypes
```

### Calcul du khwperday

```python
df['kwhPerDay'] = (df['KWH_PERIOD']/df['daysPeriod']).round(decimals = 2)
```

```python
df
```
